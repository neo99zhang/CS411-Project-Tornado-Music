```python
#### building the billboard dataset ####
import billboard
year_list=range(2006,2021)
song_name_list = []
musician_list = []
search_list = []
def get_song_from_year(year):
    chart = billboard.ChartData('alternative-songs',year=2006)
    for i in range(len(chart)):
        song = chart[i]
        song_name = str(song).split("'")[1]
        musician = str(song).split("'")[2].split('by ')[-1]
        song_name_list.append(song_name)
        musician_list.append(musician)
        search_list.append(song_name+' '+musician) 

for year in year_list:
    get_song_from_year(year)
```

```python
output_dict={}
for i in range(len(search_list)):
    try:
        search_song(search_list[i],song_name_list[i],musician_list[i])
    except:
        time.sleep(10)
        search_song(search_list[i],song_name_list[i],musician_list[i])
```
