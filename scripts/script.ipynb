```python
### ugly codes ###
import time
from selenium import webdriver
import pandas
import numpy as np
```

```python
##### A crawler #####
def search_song(search_name):
    print("begin searching! search name: "+search_name)
    url =  "https://odesli.co/"
    driver = webdriver.Chrome()
    driver.get(url=url)
    #driver.maximize_window()
    input = driver.find_element_by_id("search-page-downshift-input")
    input.send_keys(search_name)
    # key = driver.find_element_by_id("su")
    # key.click()
    time.sleep(0.5)
    # input2 = driver.find_element_by_id("search-page-downshift-item-0")
    # input2=driver.find_elements_by_css_selector('search-page-downshift-menu')
    input2 = driver.find_element_by_id("search-page-downshift-item-0")
    input3 = input2.find_element_by_css_selector('a')
    input4 = input3.find_elements_by_css_selector('div')
    count=0
    for element in input4:
        if count==4:
            input5 = element
    #     print(element.get_attribute('outerHTML'))
    #     print('\n')
        count+=1
    input5.get_attribute('outerHTML')
    input6_list = input5.find_elements_by_css_selector('button')
    count=0
    for element in input6_list:
        if count==1:
            element.click()
        count+=1
    all_handles = driver.window_handles
    mainWindow = driver.current_window_handle
    for handle in all_handles:
        driver.switch_to.window(handle)
        if search_name in handle:
            break
    input7 = driver.find_element_by_id("main-app-element")
    input8 = input7.find_element_by_css_selector('div')
    input9 = input8.find_element_by_css_selector('main')
    input10_list = input9.find_elements_by_css_selector('div:nth-child(2)')
    count=0
    for element in input10_list:
        if count==4:
            input11 = element
            break
    #         print(element.get_attribute('outerHTML'))
    #         print('\n')
        count+=1
    input11_text = input11.get_attribute('outerHTML')
    source_count = len(input11_text.split('href'))-1
    source_dict = {}
    source_full_dict = {}
    source_count_number_list=range(1,source_count+1)
    for i in source_count_number_list:
        input12 = input11.find_element_by_css_selector('a:nth-child('+str(i)+')')
        input12_text = input12.get_attribute('outerHTML')
        input12_url = input12_text.split(' ')[1].split('href=')[1]
        input12_source = input12_text.split('</div>')[0].split('>')[-1]
        source_dict[input12_source]=input12_url
    
    # print(input11.get_attribute('outerHTML'))

    source_full_list = ['Apple Music','Spotify','YouTube','YouTube Music','Pandora','Deezer','SoundCloud','Amazon Music','TIDAL','Napster','Yandex']
    for i in source_full_list:
        if i in source_dict:
            source_full_dict[i] = source_dict[i]
        else:
            source_full_dict[i] = 'not found'
        



    # for element in input12_list:
    #     print(element.get_attribute('outerHTML'))
    #     print('\n')
    #main-app-element > div > main > div:nth-child(2) > div.css-jnwrwc > a:nth-child(9) > div > svg > path:nth-child(2)
    #main-app-element > div > main > div:nth-child(2) > div.css-jnwrwc > a:nth-child(1)
    # driver.switch_to.window(mainWindow)
    #search-page-downshift-item-0 > a > div.css-1khvdun.euwaufo0 > button

    #main-app-element > div > main > div:nth-child(2) > div.css-jnwrwc > a:nth-child(2)

    #search-page-downshift-item-0 > a > div.css-1khvdun.euwaufo0 > button
    # key.click()

    #main-app-element > div > main > div:nth-child(2) > div.css-jnwrwc > a:nth-child(5) > div > svg > circle
    # input2= driver.find_element_by_id("search-page-downshift-label")
    # key.click()
    # time.sleep(5)
    driver.quit()
    print('finish current search')
    return source_full_dict
```
